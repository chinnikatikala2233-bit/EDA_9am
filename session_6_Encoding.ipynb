- **fit transform**

  - **fit**: means find the logic

    - in above example fit means create **mapper** or **dict**

    - apply the **mapper on the data**

**OneHotEncoder**

- First how many lables are there in a column

- That many new columns will be created extra

- For example case status has two lables

  - Certified

  - Denied

- So it will create two new wxtra columns

  - case_status_Certified

  - case_status_Denied

- At a time one column value become 1 other column value become 0

- This means a **one hot encoder**

|case_status|case_status_Certified|case_status_Denied|
|-|-|-|
|Certified|1|0|
|Denied|0|1|
|Denied|0|1|
